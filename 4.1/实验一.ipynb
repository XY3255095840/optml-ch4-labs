# 实验一：梯度与函数最小值探索

## 知识背景回顾

在我们使用的《最优化方法与机器学习》教材中，**梯度（Gradient）** 是一个贯穿全书的核心概念。为什么我们如此关注梯度呢？

教材中提到，梯度是**导数**这一概念在多维空间中的自然推广。对于一个单变量函数 $f(x)$，它在某一点的梯度就是它的导数 $f'(x)$。而对于一个多变量函数 $f(\mathbf{x})$，其中 $\mathbf{x}=(x_1, x_2, ..., x_n)$，它的梯度 $\nabla f(\mathbf{x})$ 是一个包含了所有偏导数的向量：

$$
\nabla f(\mathbf{x}) = \left( \frac{\partial f}{\partial x_1}, \frac{\partial f}{\partial x_2}, \dots, \frac{\partial f}{\partial x_n} \right)^T
$$

梯度的关键作用在于它**指向了函数值上升最快的方向**，而它的**负方向** `-∇f(x)` 则指向了**函数值下降最快的方向**。这正是“梯度下降法”等一系列优化算法的核心思想来源。

此外，在教材的 **定理2.39（一阶最优性条件）** 是等价于**定理4.1（无约束优化问题解的必要条件）**的：对于一个可微的凸函数，点 $\mathbf{x}_0$ 是全局最小值的点，当且仅当该点的梯度为零，即 $\nabla f(\mathbf{x}_0) = \mathbf{0}$。这个条件告诉我们，寻找最小值的过程，本质上就是在寻找梯度为零的点。

## 实验说明

1.  **一维梯度可视化（全局 vs 局部最优）**：
    * 我们选用了一个四次多项式函数，其公式为：
      $$f(x) = 0.25x^4 - \frac{2}{3}x^3 - 2.5x^2 + 6x$$
      这个函数拥有两个深度不同的“山谷”（局部最小值）。其中一个“山谷”比另一个更深，那个最深的点就是**全局最小值**。
    * 你可以通过滑杆选择函数上的任意一点。
    * 程序会动态地绘制切线、显示梯度，并用不同标记标出**局部最小值**和**全局最小值**。
    * **观察目标**：
        * 拖动滑杆，确认在两个不同的局部最小值点处，梯度值都为0。
        * **思考关键问题**：这两个局部最小值中，哪一个是全局最小值？如果一个优化算法（比如梯度下降法）从右边的“山坡”开始，它很有可能会停在右边的那个“较浅”的局部最小值点，而错过了左边那个“更深”的全局最小值点。这就是算法陷入“局部最优”的直观体现。

2.  **二维梯度可视化（标记最优值）**：
    * 我们通过叠加两个不同的负高斯函数来构造一个具有两个“碗”的曲面，其公式为：
      $$f(x, y) = -3e^{-\frac{(x-1.5)^2 + (y-1.5)^2}{2}} - 2e^{-\frac{(x+1.5)^2 + (y+1.5)^2}{3}}$$
      这两个“碗”的深度和位置都不同，从而形成了一个全局最小值和一个局部最小值。
    * 在可交互的三维曲面图上，我们**标记**出了**局部最小值点**（绿色圆点）和**全局最小值点**（红色方标）。
    * **观察目标**：旋转和缩放三维图形，从不同角度观察这两个最低点的位置和深度差异。这直观地展示了在多维空间中，优化算法同样面临着可能收敛到不同局部最优解的问题。

```python
# 导入所有需要的库
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, FloatSlider
import plotly.graph_objects as go
plt.rcParams['font.sans-serif'] = ['SimHei']  # 正确显示中文标签
plt.rcParams['axes.unicode_minus'] = False    # 正确显示负号
```

```python
# 1. 定义一个新的、拥有不同深度局部最小值的函数
def f1_global(x):
    """一个有两个不同深度局部最小值的函数"""
    return 0.25 * x**4 - (2/3) * x**3 - 2.5 * x**2 + 6 * x

def grad_f1_global(x):
    """f1_global函数的导数"""
    return x**3 - 2 * x**2 - 5 * x + 6

# 2. 寻找驻点
# 通过解 x^3 - 2x^2 - 5x + 6 = 0 得到驻点 x = -2, 1, 3
# 通过二阶导数 f''(x) = 3x^2 - 4x - 5 判断
# f''(-2) = 15 > 0 -> 局部最小值
# f''(1) = -6 < 0 -> 局部最大值
# f''(3) = 10 > 0 -> 局部最小值
local_minima_x = [-2, 3]
local_minima_y = [f1_global(x) for x in local_minima_x]

# 找出全局最小值
global_min_idx = np.argmin(local_minima_y)
global_min_x = local_minima_x[global_min_idx]
global_min_y = local_minima_y[global_min_idx]

# 3. 创建交互式绘图函数
def plot_1d_gradient_global(x_val):
    x_range = np.linspace(-3.5, 4.5, 400)
    y_range = f1_global(x_range)
    y_val = f1_global(x_val)
    grad_val = grad_f1_global(x_val)
    tangent_y = grad_val * (x_range - x_val) + y_val

    plt.figure(figsize=(10, 6))
    plt.plot(x_range, y_range, label='$f(x)=0.25x^4 - \\frac{2}{3}x^3 - 2.5x^2 + 6x$')
    plt.plot(x_range, tangent_y, '--', color='gray', label='切线 (梯度)')
    plt.plot(x_val, y_val, 'ko', markersize=8, label=f'当前点 x={x_val:.2f}')
    
    # 标记所有的局部最小值点
    for i, x_min in enumerate(local_minima_x):
        if x_min == global_min_x:
            # 单独标记全局最小值
            plt.plot(x_min, local_minima_y[i], 'r*', markersize=15, label=f'全局最小值 (x={x_min})')
        else:
            # 标记其他局部最小值
            plt.plot(x_min, local_minima_y[i], 'go', markersize=10, label=f'局部最小值 (x={x_min})')

    plt.title(f'梯度可视化：在 x = {x_val:.2f} 处的梯度为 {grad_val:.2f}', fontsize=16)
    plt.xlabel('x')
    plt.ylabel('f(x)')
    plt.legend()
    plt.grid(True)
    plt.ylim(-15, 15)
    plt.show()

# 4. 创建滑杆并绑定绘图函数
interact(plot_1d_gradient_global, x_val=FloatSlider(min=-3.5, max=4.5, step=0.1, value=4.0, description='选择x值'))

```

```python
# --- 实验二：二维函数曲面可视化（最终版） ---

# 1. 定义一个有两个“碗”的二维函数
def f2_final(x, y):
    """由两个负高斯函数叠加而成，形成两个凹陷区域"""
    term1 = -3 * np.exp(-((x - 1.5)**2 + (y - 1.5)**2) / 2) # 更深的碗
    term2 = -2 * np.exp(-((x + 1.5)**2 + (y + 1.5)**2) / 3) # 较浅且更宽的碗
    return term1 + term2

# 2. 生成网格数据
x_2d = np.linspace(-5, 5, 100)
y_2d = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x_2d, y_2d)
Z = f2_final(X, Y)

# 3. 定义局部和全局最小值点
# 理论上的最小值点就在高斯函数的中心
global_min_2d_x, global_min_2d_y = 1.5, 1.5
global_min_2d_z = f2_final(global_min_2d_x, global_min_2d_y)

local_min_2d_x, local_min_2d_y = -1.5, -1.5
local_min_2d_z = f2_final(local_min_2d_x, local_min_2d_y)


# 4. 创建曲面图
fig = go.Figure(data=[go.Surface(z=Z, x=X, y=Y, colorscale='viridis', opacity=0.8)])

# 5. 在曲面图上添加标记点
# 添加全局最小值点 (红色方标)
fig.add_trace(go.Scatter3d(
    x=[global_min_2d_x], y=[global_min_2d_y], z=[global_min_2d_z],
    mode='markers',
    marker=dict(
        size=4,
        color='red',
        symbol='diamond'
    ),
    name='全局最小值'
))

# 添加局部最小值点 (绿色圆点)
fig.add_trace(go.Scatter3d(
    x=[local_min_2d_x], y=[local_min_2d_y], z=[local_min_2d_z],
    mode='markers',
    marker=dict(
        size=4,
        color='lime',
        symbol='circle'
    ),
    name='局部最小值'
))


# 6. 设置图形布局
fig.update_layout(
    title={
        'text': '双谷函数',
        'y':0.9,
        'x':0.5,
        'xanchor': 'center',
        'yanchor': 'top'
    },
    scene=dict(
        xaxis_title='x轴',
        yaxis_title='y轴',
        zaxis_title='f(x, y)'
    ),
    autosize=True,
    margin=dict(l=65, r=50, b=65, t=90)
)

# 7. 显示图形
fig.show()
```
